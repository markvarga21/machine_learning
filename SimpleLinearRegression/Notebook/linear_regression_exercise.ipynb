```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
data = pd.read_csv("/home/mark/egyetem/4.felev/Gépi tanulás/practice/SimpleLinearRegression/exercise_dataset.csv")
```

```python
%matplotlib inline
plt.scatter(data.year, data.income)
```

```python
year = np.array(data.year).reshape(-1, 1)
income = np.array(data.income)
```

```python
reg = LinearRegression()
reg.fit(year, income)
```

```python
plt.scatter(data.year, data.income, color='red', marker='+')
plt.xlabel("YEAR")
plt.ylabel("INCOME")
plt.plot(year, reg.predict(year))
# the prediction
plt.scatter([2020],reg.predict([[2020]]), color='green')
```

```python
print(f'Income prediciont for the year of 2020 = {reg.predict([[2020]])}')
```
